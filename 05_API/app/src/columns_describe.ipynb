```python
import pandas as pd

dataset = pd.read_csv('get_around_pricing_project_clean.csv')
```

```python
for column in dataset.columns:
    print(column)
    if dataset[column].dtypes == "object":
        # Quantitative Values
        print(dataset[column].unique())
        print(f'first = {dataset[column].mode().values[0]}')
    else:
        # Qualitative Values
        print(type(dataset[column][0]))
        print(dataset[column].median())
    print()
```
